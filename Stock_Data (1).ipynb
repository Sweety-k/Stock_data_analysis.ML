# **IMPUTING MEAN FOR HANDLING MISSING VALUES**

```python
stock = data_cleaned.fillna(df.mean())
stock.isnull().sum()
```

```python
stock.info()
```

# **REMOVING OUTLIERS**

```python
Q1 = stock.quantile(0.25)
Q3 = stock.quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 0.5 * IQR
upper_bound = Q3 + 0.5 * IQR

stock = stock[~((stock >= lower_bound) & (stock <= upper_bound)).all(axis=1)]
stock
```

# **CORRELATION ANALYSIS**

```python
# Create a correlation matrix of stock returns
# Check if 'Date' is already the index
if 'Date' not in stock.columns:
    # If 'Date' is not a column, it might be the index, so reset it
    stock = stock.reset_index()

# Now convert 'Date' to datetime and set as index
stock['Date'] = pd.to_datetime(stock['Date'])
stock.set_index('Date', inplace=True)
daily_returns = stock.pct_change().dropna()
daily_returns
```

# **TIME SERIES DECOMPOSITION**

**Choose one stock and perform time series decomposition.**





```python
from statsmodels.tsa.seasonal import seasonal_decompose
stock_name = 'HDFCBANK'
stock_data = stock[stock_name]
result = seasonal_decompose(stock_data, model='additive', period= 4)
result.plot()
plt.show()
```

```python
plt.figure(figsize=(12, 8))

#Plotting the original stock price for HDFCBANK
plt.subplot(411)
plt.plot(stock_data, label='Original Stock Price')
plt.legend(loc='upper left')

#Plotting the Trend Components for HDFCBANK
plt.subplot(412)
plt.plot(result.trend, label='Trend')
plt.legend(loc='upper left')

#Plotting the Seasonal Components for HDFCBANK
plt.subplot(413)
plt.plot(result.seasonal, label='Seasonal')
plt.legend(loc='upper left')

#Plotting the Residual Components for HDFCBANK
plt.subplot(414)
plt.plot(result.resid, label='Residual')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

The time series decomposition you provided for HDFC Bank's stock price separates the data into three components: trend, seasonal, and residual. Here are the insights from each component:

**1.Trend Component:**

*   The trend component shows a clear upward trajectory from the mid-1990s to 2024.
*   There are noticeable periods of accelerated growth, particularly around 2016 and onward, with some volatility around 2020-2021.
*   This indicates a long-term positive trend in HDFC Bank's stock price, suggesting consistent growth over the years.

**2.Seasonal Component:**

*   The seasonal component appears to be a flat line around zero.
*   This suggests that there is no significant seasonal pattern in the stock price data. In other words, the stock price does not exhibit regular, repeating patterns within any given year.

**Residual Component:**

*   The residual component shows the noise or irregularities in the data after removing the trend and seasonal components.
*   There are some notable spikes in the residuals around 2011-2012 and a noticeable increase in volatility starting around 2020.

*   This suggests that there were some unusual or unexpected events affecting the stock price during these periods. The increased volatility in recent years could be attributed to market uncertainties, possibly due to economic factors or specific events impacting HDFC Bank.

**Insights from the Decomposition:**

Overall, the decomposition suggests that HDFC Bank's stock has shown strong growth over the years, with some periods of increased volatility but no significant seasonal patterns.








# **ANAMOLY DETECTION**

**Develop a method to detect anomalous price movements**

```python
def calculate_z_scores(stock):
    mean = np.mean(stock)
    std = np.std(stock)
    z_scores = (stock - mean) / std
    return z_scores

z_scores = stock.apply(calculate_z_scores)
z_scores
```

```python
threshold = 3
anomalies = (z_scores > threshold) | (z_scores < -threshold)
anomaly_summary = anomalies.sum()
print(anomaly_summary)
```

**Identify and list the top 5 most significant anomalies in the dataset**

```python
z_scores_sum = z_scores.abs().sum()

# Sort and identify top 5 most significant anomalies
top_anomalies = z_scores_sum.sort_values(ascending=False).head(5)

print("Top 5 Most Significant Anomalies:")
print(top_anomalies)
```

*   AUROPHARMA, BPCL, NATCOPHARM, EICHERMOT, ITC: These stocks show significant anomalies, suggesting unusual price movements.
